```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 200
pd.options.display.max_rows = 200
```

```python
df = pd.read_csv("./fr-en-annuaire-education.csv", sep=";", low_memory=False)
df
```

```python
# Insitutions with Public/Private status being NaN, are all Administratif services and not educational insititutions, they are thus removed from the dataset.
print("Insitutions with Public/Private status being NaN, are all of type: ", df[df["statut_public_prive"].isnull()]["type_etablissement"].unique())
df = df.loc[df['statut_public_prive'].notnull()]
df["statut_public_prive"].unique()
```

```python
# Drop insititutions with no websites
print("Insitutions with no website, are all of type: ", df[df["web"].isnull()]["type_etablissement"].unique())
# We notice no higher secondary institution is present in here, we thus remove those entries without a website.
df = df.dropna(subset=["web"])

# We drop all colunmns without importance to us.
# df = df.drop(columns=["voie_generale","voie_technologique","voie_professionnelle","telephone","fax", "code_departement", "code_academie", "identifiant_de_l_etablissement","adresse_1", "adresse_2", "adresse_3", "mail", "restauration", "hebergement", "ulis", "apprentissage", "segpa", "section_arts", "section_cinema", "section_theatre", "section_sport", "section_internationale", "section_europeenne", "lycee_agricole", "lycee_militaire", "lycee_des_metiers", "post_bac", "appartenance_education_prioritaire", "greta", "siren_siret"])
df = df[["nom_etablissement", "type_etablissement", "statut_public_prive", "web", "libelle_region", "etat", "libelle_nature", "etablissement_mere"]]
df
```

```python
# Remove institutions which closed down.
df = df.loc[df['etat'] == "OUVERT"]
df.drop(columns=["etat"], inplace=True)
df
```

```python
df["type_etablissement"].unique()
```

```python
df = df.groupby(["ENS site web", "ENS région"]).apply(lambda group: ','.join(group['ENS statut'])).reset_index(name='Finanacial Status')
df
```

```python
alcounter = 0

def f(x):
    global alcounter
    list_status = x.split(",")
    init = list_status[0]
    for i in range(1, len(list_status)):
        if list_status[i] != init:
            alcounter += 1
            print(x)
            break

    return x.split(",")[0]
    
df["Finanacial Status"] = df["Finanacial Status"].apply(lambda x: f(x))
df.rename(inplace=True, columns={"ENS site web": "website", "ENS région": "region"})
df.to_csv("./dataset-onisep-sieved.csv")
df

print(alcounter)
# consider only education level above 6
```
